```python
'''TASK 1'''
# For storing the data creating a empty nested list named as record
record = [[], [], []] 

# patients will be added only if any bed is empty
# taking index of list record and naming indexes as bed_1, bed_2, bed_3 respectively
# as it is a nested list program will take bed_1 as list of 0th index of record list
# take inputs in a single line of code as in bed_1 and and using for loop insert values in record list 
# adding one in range in for loop because in range for 0 to i range should be (i+1) 

'''TASK 2'''
def addPatient():
    if not record[0]:
        for i in range(len(record) + 1):
            bed_1 = [int((input("Enter Patient Age for Bed One: "))), str(input("Enter Patient Name for Bed One: ")), binary_emergency()]
            record[0]= bed_1
            break

    elif record[0] != 0:
        print("Bed One is occupied")
    
    back_button_1 = int(input("Enter 2 to add patient to Bed 2 or press 1 to go back to menu: "))
    if back_button_1 == 1:
        menu()
        
    else:
        if not record[1]:
            for i in range(len(record) + 1):
                bed_2 = [int((input("Enter Patient Age for Bed Two: "))), str(input("Enter Patient Name for Bed Two: ")), binary_emergency()]
                record[1]= bed_2
                break
        elif record[1] != 0:
            print("Bed Two is occupied")
        back_button_1 = int(input("Enter 2 to add patient to Bed 3 or press 1 to go back to menu:  "))
        if back_button_1 == 1:
            menu()
        else:
            if not record[2]:
                for i in range(len(record) + 1):
                    bed_3 = [int((input("Enter Patient Age for Bed Three: "))), str(input("Enter Patient Name for Bed Three: ")), binary_emergency()]
                    record[2] = bed_3
                    break
            elif record[2] != 0:
                print("Bed Three is occupied")
            back_button_1 = int(input("Enter 1 to go back to menu: "))
            if back_button_1 == 1:
                menu()
            else:
                menu()



# Defined emergency as a function to take input as yes or no only
# This function is used while taking inputs in add patient

def binary_emergency():
    Emergency = input("Emergency (yes or no): ")
    if Emergency == 'yes':
        return 'yes'
    elif Emergency == 'no':
        return 'no'
    else:
        print("Only Enter yes or no")
        binary_emergency()

        
'''TASK 3'''
# shows patients data 
# it is using a for loop which print list indexes in vertical form
# using format function to print age name and emergency with a space of 6 between thwm

def showPatients():
        print("{:<6} {:<6} {:<6}".format('Age', 'Name', 'Emergency'))
        for x in range(len(record)):
            print(record[x])
        
        back_button_1 = int(input("Enter 1 to go back to menu: "))
        if back_button_1 == 1:
            menu()
        else:
            showPatients()
        

# the function check_bed can check which bed is occupied or unoccupied immediately
# it is using simple if - elif - else loop to mark beds
# using of not to check value inside a particular index to be 0 
# program marks bed unoccupied if value of that index is 0 and any other value marks it as occupied

def check_bed_status():
    if not record[0]:
        print("bed one is unoccupied")
    elif record[0] != 0:
        print("bed one is occupied")

    if not record[1]:
        print("bed two is unoccupied")
    elif record[1] != 0:
        print("bed two is occupied")

    if not record[2]:
        print("bed three is unoccupied")
    elif record[2] != 0:
        print("bed three is occupied")

    back_button_1 = int(input("Enter 1 to go back to menu: "))
    if back_button_1 == 1:
        menu()
    else:
        check_bed_status()
        
    
'''TASK 4'''
# Release patient by input bed no 
# subtracting one because the list index will be one less from bed number as for bed 1 ,2 ,3 index will be 0 ,1 ,2 respectively
# Using insert function to add 0 in place of released paitent to maintain list index of 2
# inserting 0 as in check bed status function 0 is assigned to mark bed as unoccupied

def releasePatient():
    x = (int(input("Enter Bed Number To Release Patient: ")) - 1)
    del record[x]
    record.insert(x, 0)   
    back_button_1 = int(input("Enter 1 to go back to menu: "))
    if back_button_1 == 1:
        menu()
    else:
        releasePatient()

        

'''TASK 5'''
# Defined a menu function to make program easy to navigate and recurring
# Binded some keys to select a particular function using if - elif loop 
# Key is assigned to a funtion based on its order in menu function
# To exit used return 0 to terminate program
def menu():
    print("1 Check Bed Status\n"
          "2 Add Patient\n"
          "3 Release Patient\n"
          "4 Display Patient Data\n"
          "5 Exit Program")
    select = int(input("Select Option: "))
    if select == 1:
        check_bed_status()
    elif select == 2:
        addPatient()
    elif select == 3:
        releasePatient()
    elif select == 4:
        showPatients()
    elif select == 5:
        return 0
    else:
        print("Enter Correct Option")
        menu()

        
# Intialisation of program
menu()
```
